```python
import numpy as np
import pandas as pd
import pandas_datareader.data as web
import matplotlib.pyplot as plt
```

```python
goog = web.DataReader('GOOG' , data_source = 'morningstar', start= '4/1/2013' , end = '3/25/2018')
goog.tail()
```

```python
goog['Log_Ret'] = np.log(goog['Close'] / goog['Close'].shift(1))
goog['Volatility'] = pd.Series(goog['Log_Ret']).rolling(window=252).std() * np.sqrt(252)
%matplotlib inline
goog[['Close', 'Volatility']].plot(subplots=True, color='blue',
figsize=(15, 10))
```
