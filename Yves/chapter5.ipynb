```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#1D dataset

#gen 20 statndard normally distributed (pseudo) random  numbers

np.random.seed(1000)
y = np.random.standard_normal(20)
```

```python
x = range(len(y))
```

```python
plt.plot(x,y)
```

```python
plt.plot(y)
```

```python
plt.plot(y.cumsum())
plt.grid(True)
#plt.axis('tight')
plt.xlim(-1 , 20)
plt.ylim(np.min(y.cumsum()) - 1 , np.max(y.cumsum())+1)
```

```python
plt.figure(figsize= (7,4))
plt.plot(y.cumsum(), 'y',lw = 1.5)
plt.plot(y.cumsum(), 'D')
plt.grid(True)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
```

```python
# 2D data set


# numpy array , shape : 20 * 2 
np.random.seed(2000)
y = np.random.standard_normal((20,2)).cumsum(axis = 0)
```

```python
# 2 separate subplots

plt.figure(figsize=(7,5))
plt.subplot(211)
plt.plot(y[:,0], lw = 1.5 , label = '1st')
plt.plot(y[:,0] , 'ro')
plt.grid(True)
plt.legend(loc = 0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A simple plot')

plt.subplot(212)
plt.plot(y[:,1], 'g' ,lw = 1.5 , label = '1st')
plt.plot(y[:,1] , 'ro')
plt.grid(True)
plt.legend(loc = 0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
```

```python
plt.figure(figsize=(9,4))
plt.subplot(121)
plt.plot(y[:,0], lw = 1.5 , label = '1st')
plt.plot(y[:,0] , 'ro')
plt.grid(True)
plt.legend(loc = 0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('1st Data set')

plt.subplot(122)
plt.bar(np.arange(len(y)) , y[:,1],width = 0.5,color = 'g', label = '2nd')
plt.grid(True)
plt.legend(loc = 0)
plt.axis('tight')
plt.xlabel('index')
plt.title('2nd Data set')
```

```python
# Other plot styles

#scatter plot

y = np.random.standard_normal((1000,2))
plt.figure(figsize=(7,5))
plt.plot(y[:,0] , y[:,1],'ro')
plt.grid(True)
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot')
```

```python
from matplotlib.patches import Polygon
```

```python
def func(x):
    return 0.5*np.exp(x) + 1
```

```python
a,b = 0.5,1.5 #limits of integration
```

```python
x = np.linspace(0,2)
y=func(x)
```

```python
fig,ax = plt.subplots(figsize = (7,5))
plt.plot(x,y,'b',linewidth = 2)
plt.ylim(ymin=0)

Ix = np.linspace(a,b)
Iy = func(Ix)
verts = [(a,0)] + list(zip(Ix,Iy)) + [(b,0)]
poly = Polygon(verts,facecolor = '0.7' , edgecolor = '0.5')
ax.add_patch(poly)

plt.text(0.5*(a+b),1,r"$\int_a^b f(x)\mathrm{d}x$" , horizontalalignment='center',fontsize=20)

plt.figtext(0.9,0.075,'$x$')
plt.figtext(0.075,0.9,'$f(x)$')
ax.set_xticks((a,b))
ax.set_xticklabels(('$a$' ,'$b$'))
ax.set_yticks([func(a),func(b)])
ax.set_yticklabels(('$f(a)$' , '$f(b)$'))
plt.grid(True)
```
